```python
import yfinance as yf
import pandas as pd

eurusd_daily = yf.download("EURUSD=X", start="2010-01-01", interval="1d")

print(f"Filas descargadas: {len(eurusd_daily)}")
print(f"Período: {eurusd_daily.index[0]} → {eurusd_daily.index[-1]}")
print(eurusd_daily.tail())
```

```python
# Descargar EUR/USD en 4H
eurusd_4h = yf.download("EURUSD=X", start="2020-01-01", interval="4h")

print(f"Filas descargadas: {len(eurusd_4h)}")
print(f"Período: {eurusd_4h.index[0]} → {eurusd_4h.index[-1]}")
print(eurusd_4h.tail())
```

```python
# Descargar EUR/USD en 4H (máximo ~2 años en yfinance)
eurusd_4h = yf.download("EURUSD=X", start="2024-01-01", interval="4h")

print(f"Filas descargadas: {len(eurusd_4h)}")
print(f"Período: {eurusd_4h.index[0]} → {eurusd_4h.index[-1]}")
print(eurusd_4h.tail())
```

```python
eurusd_4h = yf.download("EURUSD=X", start="2024-01-01", interval="4h")

if len(eurusd_4h) > 0:
    print(f"Filas descargadas: {len(eurusd_4h)}")
    print(f"Período: {eurusd_4h.index[0]} → {eurusd_4h.index[-1]}")
    print(eurusd_4h.tail())
else:
    print("Sin datos — probando período más corto")
    eurusd_4h = yf.download("EURUSD=X", period="60d", interval="4h")
    print(f"Filas descargadas: {len(eurusd_4h)}")
    print(eurusd_4h.tail())
```

```python
# Guardar datos en disco
eurusd_daily.to_csv("../data/raw/eurusd_daily.csv")
eurusd_4h.to_csv("../data/raw/eurusd_4h.csv")

print("Datos guardados:")
print(f"  eurusd_daily.csv — {len(eurusd_daily)} filas")
print(f"  eurusd_4h.csv   — {len(eurusd_4h)} filas")
```
